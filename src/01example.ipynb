```python
import numpy as np
import matplotlib.pyplot as plt
from interpolation import natural_cubic_spline, conditional_cubic_spline, eval_spline
```

```python
X = [1, 2, 3, 4, 5]
Y = np.exp(X)
CS = conditional_cubic_spline(X, Y, 100, -100)
CS1 = natural_cubic_spline(X, Y)
I = np.linspace(min(X), max(X), 100)
plt.plot(I, eval_spline(CS, X, I))
plt.plot(I, eval_spline(CS1, X, I))
plt.scatter(X, Y)
```
