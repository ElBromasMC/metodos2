```python
import numpy as np
import matplotlib.pyplot as plt
from interpolation import natural_cubic_spline, conditional_cubic_spline, eval_spline
```

```python
f = lambda x : np.exp(x)

X = [0, 1, 2, 3]
Y = f(X)
CS = natural_cubic_spline(X, Y)
CS1 = conditional_cubic_spline(X, Y, 1, np.exp(3))
I = np.linspace(min(X), max(X), 100)
R = eval_spline(CS, X, I)
R1 = eval_spline(CS1, X, I)
plt.plot(I, R, label="natural")
plt.plot(I, R1, label="conditional")
plt.legend()
plt.scatter(X, Y)
```

```python
E = np.abs(R - f(I))
E1 = np.abs(R1 - f(I))
plt.plot(I, E, label="natural")
plt.plot(I, E1, label="conditional")
plt.legend()
```

```python
print("Natural max error:", max(E))
print("Conditional max error:", max(E1))
print(CS[2].deriv().deriv()(3))
```
