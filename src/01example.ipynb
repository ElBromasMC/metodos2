# Ejercicio 1

Aproximamos la función exponencial $e^x$ usando el spline lineal y cúbico (natural y condicional)


```python
import numpy as np
import matplotlib.pyplot as plt
from interpolation import natural_cubic_spline, conditional_cubic_spline, lineal_spline, eval_spline
```

## Gráfico de las interpolaciones

```python
f = lambda x : np.exp(x)

X = [0, 1, 2, 3]
Y = f(X)

CS = natural_cubic_spline(X, Y)
CS1 = conditional_cubic_spline(X, Y, 1, np.exp(3))
CS2 = lineal_spline(X, Y)

I = np.linspace(min(X), max(X), 100)

R = eval_spline(CS, X, I)
R1 = eval_spline(CS1, X, I)
R2 = eval_spline(CS2, X, I)

plt.plot(I, R, label="natural")
plt.plot(I, R1, label="conditional")
plt.plot(I, R2, label="lineal")
plt.legend()
plt.scatter(X, Y)
plt.show()
```

## Gráfico del error de las interpolaciones

```python
E = np.abs(R - f(I))
E1 = np.abs(R1 - f(I))
E2 = np.abs(R2 - f(I))

plt.plot(I, E, label="natural")
plt.plot(I, E1, label="conditional")
plt.plot(I, E2, label="lineal")
plt.legend()
plt.show()
```
